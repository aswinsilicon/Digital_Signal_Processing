```python

import numpy as np
import matplotlib.pyplot as pplot

#define unit: unit impulse
n= np.arange(-20,20)
x= np.where(n==0,1,0)

#initialize output
y=np.zeros_like(n,dtype=float)

#difference equation used:
for i in range(0,len(n)):
    y[i] = 0.5*y[i-1] + x[i]
    
pplot.stem(n,y)
pplot.stem(n,x,'r')
pplot.xlabel("n")
pplot.ylabel("convolution impulse function")
pplot.title("convolution impulse with python function")
pplot.show()
    
```

```python

import numpy as np
import matplotlib.pyplot as pplot

#define unit: unit step
n= np.arange(-20,20)
fc=50 #critical frequency
f=2   #selected frequency
sine=np.sin(2*np.pi*n*(f/fc))

#initialize output
y=np.zeros_like(n,dtype=float)

#difference equation used:
for i in range(1,len(n)):
    y[i] = 0.5*y[i-1] + sine[i]
    
pplot.stem(n,y)

pplot.stem(n,sine,'r')
pplot.xlabel("n")
pplot.ylabel("convolution sinusoidal function")
pplot.title("convolution sinusoidal with python function")
pplot.show()
```

```python
import numpy as np
import matplotlib.pyplot as pplot

#define unit: unit step
n= np.arange(-20,20)
step= np.where(n>=0,1,0)

h=0.5**n
res_length= len(step) + len(h) -1

y=np.zeros(res_length)

for i in range(res_length):
    for k in range(len(step)):
        if 0<=(i-k)<len(h):
            y[i] = y[i] + step[k]*h[i-k]
            
ycut=y[0:40]            
pplot.stem(n,ycut,'r')
pplot.xlabel("n")
pplot.ylabel("convolution step function without")
pplot.title("linear convolution step function")
pplot.show()
```

```python
import numpy as np
import matplotlib.pyplot as pplot

n= np.arange(-20,20)
fc=50 #critical frequency
f=2   #selected frequency
sine =np.sin(0.1*np.pi*n)
noise =np.random.normal(0,0.3,size=sine.shape)
noisy_signal = sine + noise

a=0.1

#initialize output
y=np.zeros_like(sine)

#difference equation used:
for i in range(1,40):
    y[i] = ((1-a))*y[i-1] + (a*noisy_signal[i])
    
pplot.subplot(3,2,1)
pplot.stem(n,sine,'r')
pplot.subplot(3,2,2)
pplot.stem(n,noisy_signal,'g')
pplot.subplot(3,2,3)
pplot.stem(n,y)
pplot.show()
```
